## ⚡ ENERGY — Compact Horizontal Notes

**Definition:** Energy is the ability to do work or cause change.  
**In my words:** All objects and substances contain stored potential that allows transformations such as motion, heat, or reactions.

---

## Main Formula

**Energy (chemical form):**  
**E = m · LHV** → m = mass of the substance (kg), LHV = lower heating value, energy released per unit mass (J/kg).  
Example substance: biomass, hydrogen, biofuel, natural gas, food molecules.  
Example: m = 2 kg, LHV = 15 MJ/kg → **E = 30 MJ = 3.0 × 10⁷ J**.

---

## NOT Energy (common confusions)

Force = push or pull, **not stored energy**.  
Power = rate of energy use, **not energy itself**.  
Temperature = thermal state measure, **not total energy**.  
Mass alone = matter quantity without considering stored energy.

---

## Mini Exercise

A fuel sample has **m = 0.5 kg** and **LHV = 40 MJ/kg**.  
Find the released energy using **E = m · LHV**.

**Answer:** **E = 20 MJ = 2.0 × 10⁷ J**.


```python
# ⚡ Compact Chemical Energy Calculator

def chemical_energy(m, lhv):
    E_MJ = m * lhv
    E_J = E_MJ * 1e6
    return E_MJ, E_J


print("=== Chemical Energy Calculator ===")

try:
    m = float(input("Mass (kg): "))
    lhv = float(input("LHV (MJ/kg): "))

    E_MJ, E_J = chemical_energy(m, lhv)

    print("\n--- Result ---")
    print(f"Energy: {E_MJ:.3f} MJ")
    print(f"Energy: {E_J:.3e} J")

except ValueError:
    print("\nInvalid input. Please enter numeric values.")
```

## ⚡ POWER — Compact Horizontal Notes

**Definition:** Power is the rate at which energy is transferred or work is done.  
**In my words:** Power describes **how fast energy is moving or being transferred** over time.

---

## Main Formula

**Power:**  
**P = E / t** → P = power (W), E = energy (J), t = time (s).  
Meaning: power measures **energy per unit of time**.

**Example situation:**  
A device that uses **100 J** of energy in **5 s**.  
**P = 100 / 5 = 20 W**.

---

## Units

Power is measured in **watts (W)**.

**1 watt = 1 joule per second**  
**1 W = 1 J/s**

Common larger units:  
- **kW = 10³ W**  
- **MW = 10⁶ W**

---

## Mini Exercise

A machine transfers **500 J** of energy in **10 s**.  
Find the power using **P = E / t**.
