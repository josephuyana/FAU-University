# Linear Algebra

## Vector Space â€” Definition

A **vector space** is a set of objects called **vectors** where two operations are defined:

- **Vector addition**
- **Scalar multiplication** (multiplication by real numbers)

To be considered a vector space, the set must satisfy:

**Addition properties**
- Closed under addition  
- Commutative and associative  
- Contains a **zero vector**  
- Every vector has an **additive inverse**

**Scalar multiplication properties**
- Closed under scalar multiplication  
- Distributive over vector addition and scalar addition  
- Associative with scalars  
- Identity element: $1v = v$

**Common examples**
- $\mathbb{R}^2$, $\mathbb{R}^3$  
- Polynomial spaces  
- Spaces of continuous functions  


## Visualization of the Vector Space in $\mathbb{R}^3$


```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 3D figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Draw a transparent cube to represent the vector space R^3
r = [-3, 3]
for s in r:
    for e in r:
        ax.plot([s, s], [e, e], r)
        ax.plot([s, s], r, [e, e])
        ax.plot(r, [s, s], [e, e])

ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)
ax.set_zlim(-3, 3)

# Hide only the axis numbers (keep ticks for grid)
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])

# Ensure grid is visible
ax.grid(True)

ax.set_title("Vector Space $\\mathbb{R}^3$")
plt.show()
```

## Subspace â€” Definition

A **subspace** is a **subset of a vector space** that is also a **vector space under the same operations**.

A subset $W \subseteq V$ is a subspace if:

1. It **contains the zero vector**  
2. It is **closed under vector addition**  
3. It is **closed under scalar multiplication**

**Typical examples**
- A **line through the origin** in $\mathbb{R}^2$  
- A **plane through the origin** in $\mathbb{R}^3$  
- Solution sets of **homogeneous linear systems**

A line or plane **not passing through the origin** is **not** a subspace.


## Visualization of a Subspace inside $\mathbb{R}^3$


```python
import numpy as np
import matplotlib.pyplot as plt

# Create grid for a plane (subspace z = x + y through the origin)
x = np.linspace(-3, 3, 20)
y = np.linspace(-3, 3, 20)
X, Y = np.meshgrid(x, y)
Z = X + Y  # plane through origin â†’ valid subspace

# Create 3D figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Draw cube representing the full vector space R^3
r = [-3, 3]
for s in r:
    for e in r:
        ax.plot([s, s], [e, e], r, alpha=0.2)
        ax.plot([s, s], r, [e, e], alpha=0.2)
        ax.plot(r, [s, s], [e, e], alpha=0.2)

# Plot the subspace plane (highlighted)
ax.plot_surface(X, Y, Z)

ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)
ax.set_zlim(-3, 3)

# Hide only the axis numbers (keep ticks for grid)
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])

# Ensure grid is visible
ax.grid(True)

ax.set_title("Subspace inside $\\mathbb{R}^3$")
plt.show()
```

## Span â€” Definition

The **span** of a set of vectors is the set of **all linear combinations**
that can be formed from them.

If $v_1, \dots, v_k \in V$, then:

$$
\text{Span}\{v_1, \dots, v_k\}
= \{a_1 v_1 + \dots + a_k v_k \mid a_1, \dots, a_k \in \mathbb{R}\}.
$$

### Key idea
The **span is the resulting subspace** created by combining the vectors.

### Geometric intuition in $\mathbb{R}^3$
- One vector â†’ a **line through the origin**  
- Two independent vectors â†’ a **plane through the origin**  
- Three independent vectors â†’ **all of $\mathbb{R}^3$**


```python
import numpy as np
import matplotlib.pyplot as plt

# Two vectors whose span is a plane in R^3
v1 = np.array([1, 0, 0])
v2 = np.array([0, 1, 0])

# Create linear combinations to visualize the span (a plane)
a = np.linspace(-2, 2, 25)
b = np.linspace(-2, 2, 25)
A, B = np.meshgrid(a, b)

X = A * v1[0] + B * v2[0]
Y = A * v1[1] + B * v2[1]
Z = A * v1[2] + B * v2[2]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Draw the span as a surface (the RESULTING subspace)
ax.plot_surface(X, Y, Z, alpha=0.6)

ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])
ax.grid(True)

ax.set_title("Span (resulting plane) in $\\mathbb{R}^3$")
plt.show()
```

## Generating Set â€” Definition

A **generating set** is the **collection of vectors**
whose **span produces a space or subspace**.

$$
\text{If } \text{Span}(S) = W,\ \text{then } S \text{ is a generating set of } W.
$$

### Key difference from span
- The **generating set** = the **original vectors**.  
- The **span** = the **space produced by them**.

### Important fact
If the generating set has **no redundant vectors**, it becomes a **basis**.


```python
import numpy as np
import matplotlib.pyplot as plt

# Three generating vectors in R^3
v1 = np.array([1, 0, 0])
v2 = np.array([0, 1, 0])
v3 = np.array([0, 0, 1])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Show ONLY the generating vectors as arrows from the origin
ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], linewidth=2)
ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], linewidth=2)
ax.quiver(0, 0, 0, v3[0], v3[1], v3[2], linewidth=2)

ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_zlim(-2, 2)

ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])
ax.grid(True)

ax.set_title("Generating set of three vectors in $\\mathbb{R}^3$")
plt.show()
```

## Intersection of Subspaces â€” Definition

The **intersection** of two subspaces $U$ and $W$ is the set of vectors
that belong to **both** subspaces.

$$
U \cap W = \{v \mid v \in U \text{ and } v \in W\}.
$$

### Key idea
The intersection of subspaces is **always a subspace**.

### Geometric intuition in $\mathbb{R}^3$
Two planes through the origin usually intersect in a **line through the origin**.


```python
import numpy as np
import matplotlib.pyplot as plt

# Create grids for two planes
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)

Z1 = X          # Plane 1: z = x
Z2 = -X         # Plane 2: z = -x

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot both planes
ax.plot_surface(X, Y, Z1, alpha=0.3)
ax.plot_surface(X, Y, Z2, alpha=0.3)

# Intersection line (y-axis)
t = np.linspace(-2, 2, 100)
ax.plot(np.zeros_like(t), t, np.zeros_like(t), linewidth=3)

ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])
ax.grid(True)

ax.set_title("Intersection of two subspaces in $\\mathbb{R}^3$")
plt.show()
```

## Union of Subspaces

The **union** of two subspaces $U$ and $W$ is the set of vectors that
belong to **U or W**:

$$
U \cup W = \{v \mid v \in U \text{ or } v \in W\}.
$$

Geometrically, this means we take **both shapes together**
without combining their vectors.


```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plane U: z = 0
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)
Z = 0 * X
ax.plot_surface(X, Y, Z, alpha=0.3)

# Line W: z-axis
t = np.linspace(-2, 2, 100)
ax.plot(np.zeros_like(t), np.zeros_like(t), t, linewidth=3)

ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_zlim(-2, 2)
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])
ax.grid(True)

ax.set_title("Union = plane OR line (no combinations)")
plt.show()
```

## Sum of Subspaces

The **sum** of two subspaces contains **all possible sums**
of one vector from each subspace:

$$
U + W = \{u + w \mid u \in U,\; w \in W\}.
$$

Geometrically, this means we **combine directions**
to create a **larger subspace**.


```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Same plane U
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)
Z = 0 * X
ax.plot_surface(X, Y, Z, alpha=0.25)

# Same line W
t = np.linspace(-2, 2, 100)
ax.plot(np.zeros_like(t), np.zeros_like(t), t, linewidth=3)

# Resulting independent directions filling R^3
ax.quiver(0, 0, 0, 1, 0, 0, linewidth=2)
ax.quiver(0, 0, 0, 0, 1, 0, linewidth=2)
ax.quiver(0, 0, 0, 0, 0, 1, linewidth=3)

ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_zlim(-2, 2)
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])
ax.grid(True)

ax.set_title("Sum = combinations filling $\\mathbb{R}^3$")
plt.show()
```

## Why the Union is NOT a Subspace

A subspace must be **closed under addition**:

> If two vectors are in the set, their sum must also be in the set.

Take:

- $u$ in the plane  
- $w$ in the line  

Their sum leaves **both shapes**,  
so the union is **not closed under addition** â†’ **not a subspace**.


```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plane
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)
Z = 0 * X
ax.plot_surface(X, Y, Z, alpha=0.3)

# Line
t = np.linspace(-2, 2, 100)
ax.plot(np.zeros_like(t), np.zeros_like(t), t, linewidth=3)

# Vectors showing failure of closure
u = np.array([1, 1, 0])
w = np.array([0, 0, 1])
s = u + w

ax.quiver(0, 0, 0, *u, linewidth=2)
ax.quiver(0, 0, 0, *w, linewidth=2)
ax.quiver(0, 0, 0, *s, linewidth=3)  # outside union

ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_zlim(-2, 2)
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])
ax.grid(True)

ax.set_title("Sum leaves the union â†’ not closed")
plt.show()
```

## Why the Sum IS a Subspace

The sum includes **all combinations** $u + w$.

Because of this, it:

- contains the **zero vector**
- is **closed under addition**
- is **closed under scalar multiplication**

Therefore, the sum **always forms a subspace**.


```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Show three independent directions (a subspace)
ax.quiver(0, 0, 0, 1, 0, 0, linewidth=2)
ax.quiver(0, 0, 0, 0, 1, 0, linewidth=2)
ax.quiver(0, 0, 0, 0, 0, 1, linewidth=3)

ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_zlim(-2, 2)
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])
ax.grid(True)

ax.set_title("All combinations form a subspace ($\\mathbb{R}^3$)")
plt.show()
```

## When the Union IS a Subspace

If one subspace is **inside** the other,
their union equals the **larger subspace**.

In that special case, the union **is closed**
and **is a subspace**.


```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Large plane W
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)
Z = 0 * X
ax.plot_surface(X, Y, Z, alpha=0.3)

# Small line U inside the plane
t = np.linspace(-2, 2, 100)
ax.plot(t, np.zeros_like(t), np.zeros_like(t), linewidth=4)

ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_zlim(-2, 2)
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])
ax.grid(True)

ax.set_title("Union equals the larger subspace â†’ valid")
plt.show()
```

## Direct Sum of Subspaces â€” Definition

The **direct sum** of two subspaces $U$ and $W$ means that every vector
in $U + W$ can be written in a **unique way** as a sum of a vector in $U$
and a vector in $W$.

$$
U \oplus W \quad \text{if and only if} \quad U \cap W = \{0\}.
$$

### Key idea
The only vector shared by both subspaces is the **zero vector**.

### Geometric intuition in $\mathbb{R}^3$
Two lines through the origin in different directions form a **plane**
where each vector has a **unique decomposition** along the two lines.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for two lines through the origin
t = np.linspace(-2, 2, 100)

# Line U along x-axis
x1, y1, z1 = t, np.zeros_like(t), np.zeros_like(t)

# Line W along y-axis
x2, y2, z2 = np.zeros_like(t), t, np.zeros_like(t)

# Plane generated by U âŠ• W (xy-plane)
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)
Z = np.zeros_like(X)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the plane
ax.plot_surface(X, Y, Z, alpha=0.2)

# Plot the two generating lines
ax.plot(x1, y1, z1, linewidth=3)
ax.plot(x2, y2, z2, linewidth=3)

ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])
ax.grid(True)

ax.set_title("Direct Sum of two subspaces in $\\mathbb{R}^3$")
plt.show()
```

## Sum vs Direct Sum of Subspaces â€” Difference

The **sum** of two subspaces $U$ and $W$ is the set of all vectors
that can be written as

$$
U + W = \{u + w \mid u \in U,\; w \in W\}.
$$

This representation is **not necessarily unique**.

The **direct sum** $U \oplus W$ occurs when the representation
**is unique**, which happens exactly when

$$
U \cap W = \{0\},
$$

and the vectors that generate the subspaces are **linearly independent (LI)**.

### Key idea
- **Sum:** vectors may have multiple decompositions.  
- **Direct sum:** generators are **LI** and every vector has a **unique decomposition**.

### Geometric intuition in $\mathbb{R}^3$
- If two planes share a line â†’ the sum is **not direct**.  
- If two lines only meet at the origin and are **LI** â†’ the sum **is direct**.


```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 5))

# -------------------------
# Case 1: Sum but NOT direct
# -------------------------
ax1 = fig.add_subplot(121, projection='3d')

x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)

Z1 = np.zeros_like(X)   # Plane U: z = 0
Z2 = Y                  # Plane W: z = y  (share a line)

ax1.plot_surface(X, Y, Z1, alpha=0.3)
ax1.plot_surface(X, Y, Z2, alpha=0.3)

t = np.linspace(-2, 2, 100)
ax1.plot(t, np.zeros_like(t), np.zeros_like(t), linewidth=3)  # shared line

ax1.set_xticklabels([])
ax1.set_yticklabels([])
ax1.set_zticklabels([])
ax1.set_title("Sum (not direct)")
ax1.grid(True)

# -------------------------
# Case 2: Direct sum
# -------------------------
ax2 = fig.add_subplot(122, projection='3d')

t = np.linspace(-2, 2, 100)

# Two LI lines through origin (x-axis and y-axis)
x1, y1, z1 = t, np.zeros_like(t), np.zeros_like(t)
x2, y2, z2 = np.zeros_like(t), t, np.zeros_like(t)

# Plane generated by the direct sum (xy-plane)
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)
Z = np.zeros_like(X)

ax2.plot_surface(X, Y, Z, alpha=0.2)
ax2.plot(x1, y1, z1, linewidth=3)
ax2.plot(x2, y2, z2, linewidth=3)

ax2.set_xticklabels([])
ax2.set_yticklabels([])
ax2.set_zticklabels([])
ax2.set_title("Direct sum (LI generators)")
ax2.grid(True)

plt.show()
```

## Linear Independence â€” Definition

A set of vectors $\{v_1, v_2, \dots, v_k\}$ is **linearly independent (LI)**
if the only solution to

$$
c_1 v_1 + c_2 v_2 + \cdots + c_k v_k = 0
$$

is

$$
c_1 = c_2 = \cdots = c_k = 0.
$$

If a non-trivial combination gives the zero vector, the vectors are
**linearly dependent**.

### Key idea
- **LI:** no vector can be written as a combination of the others.  
- **Dependent:** at least one vector is a combination of the rest.

### Geometric intuition in $\mathbb{R}^3$
- Two vectors in different directions â†’ **linearly independent** (form a plane).  
- Two vectors in the same line â†’ **linearly dependent**.


```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 5))

# -------------------------
# Case 1: Linearly dependent vectors
# -------------------------
ax1 = fig.add_subplot(121, projection='3d')

t = np.linspace(-2, 2, 100)

# Both vectors lie on the same line
v1 = np.array([1, 0, 0])
v2 = np.array([2, 0, 0])  # multiple of v1

ax1.plot(t*v1[0], t*v1[1], t*v1[2], linewidth=3)
ax1.plot(t*v2[0], t*v2[1], t*v2[2], linewidth=3, linestyle="--")

ax1.set_xticklabels([])
ax1.set_yticklabels([])
ax1.set_zticklabels([])
ax1.set_title("Linearly dependent")
ax1.grid(True)

# -------------------------
# Case 2: Linearly independent vectors
# -------------------------
ax2 = fig.add_subplot(122, projection='3d')

v1 = np.array([1, 0, 0])
v2 = np.array([0, 1, 0])

# Plane generated by the two LI vectors
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)
Z = np.zeros_like(X)

ax2.plot_surface(X, Y, Z, alpha=0.2)
ax2.plot(t*v1[0], t*v1[1], t*v1[2], linewidth=3)
ax2.plot(t*v2[0], t*v2[1], t*v2[2], linewidth=3)

ax2.set_xticklabels([])
ax2.set_yticklabels([])
ax2.set_zticklabels([])
ax2.set_title("Linearly independent")
ax2.grid(True)

plt.show()
```

## How to Check Linear Independence in an Exam

To determine if vectors are **linearly independent (LI)**, solve

$$
c_1 v_1 + c_2 v_2 + \cdots + c_k v_k = 0.
$$

- If the **only solution** is all coefficients equal to **zero** â†’ the vectors are **LI**.  
- If there exists a **non-zero solution** â†’ the vectors are **linearly dependent**.

### Practical methods
1. **Matrix method:** place the vectors as columns of a matrix and compute the  
   **determinant** (square case) or **rank**.
   - Determinant $\neq 0$ â†’ **LI**  
   - Determinant $= 0$ â†’ **dependent**

2. **Row reduction (RREF):**
   - A pivot in every column â†’ **LI**  
   - A free variable appears â†’ **dependent**

### Key idea
Linear independence means **no vector is redundant** in the set.


```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 5))

# -------------------------
# Case 1: Determinant = 0 (dependent)
# -------------------------
ax1 = fig.add_subplot(121, projection='3d')

t = np.linspace(-2, 2, 100)

v1 = np.array([1, 1, 0])
v2 = np.array([2, 2, 0])  # multiple of v1 â†’ determinant = 0

ax1.plot(t*v1[0], t*v1[1], t*v1[2], linewidth=3)
ax1.plot(t*v2[0], t*v2[1], t*v2[2], linewidth=3, linestyle="--")

ax1.set_title("Dependent (det = 0)")
ax1.set_xticklabels([])
ax1.set_yticklabels([])
ax1.set_zticklabels([])
ax1.grid(True)

# -------------------------
# Case 2: Determinant â‰  0 (independent)
# -------------------------
ax2 = fig.add_subplot(122, projection='3d')

v1 = np.array([1, 0, 0])
v2 = np.array([0, 1, 0])
v3 = np.array([0, 0, 1])  # three independent vectors â†’ det â‰  0

for v in [v1, v2, v3]:
    ax2.plot(t*v[0], t*v[1], t*v[2], linewidth=3)

ax2.set_title("Independent (det â‰  0)")
ax2.set_xticklabels([])
ax2.set_yticklabels([])
ax2.set_zticklabels([])
ax2.grid(True)

plt.show()
```
