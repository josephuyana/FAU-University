# Renewable Energies

## ðŸ”„ Energy Flow

The energy we use every day goes through several **stages**, from nature to the final benefit we receive.

---

### ðŸŒ± Primary Energy
Energy that exists **directly in nature**, before any human transformation.

**Examples:**  
- Coal  
- Wind  
- River water  
- Solar radiation  

---

### âš¡ Secondary Energy
Energy obtained by **transforming primary energy** so it can be transported or used more efficiently.

**Examples:**  
- Electricity generated in power plants  
- Gasoline refined from oil  

---

### ðŸ”Œ Final Energy
Energy that **reaches the consumer** ready to be used.

**Examples:**  
- Electricity from a home outlet  
- Fuel inside a carâ€™s tank  

---

### ðŸ’¡ Energy Service
The **real benefit** we obtain from using energy.

**Examples:**  
- Light for visibility  
- Heat for cooking or heating  
- Motion in vehicles or machines  

---

## ðŸ§­ Visual Summary

**Nature â†’ Transformation â†’ Consumption â†’ Real Benefit**

Each stage brings energy closer to what truly matters:  
ðŸ‘‰ **improving everyday life**.


```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create 3D figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Coordinates for each stage
x = [0, 1, 2, 3]
y = [0, 0, 0, 0]
z = [0, 1, 2, 3]

labels = [
    "Primary\nEnergy",
    "Secondary\nEnergy",
    "Final\nEnergy",
    "Energy\nService"
]

# Draw flow line with markers
ax.plot(x, y, z, marker='o')

# Add labels
for i in range(len(x)):
    ax.text(x[i], y[i], z[i], labels[i], ha='center', va='bottom')

# Hide only the axis numbers (keep grid)
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])

# Remove axis titles but keep grid
ax.set_xlabel("")
ax.set_ylabel("")
ax.set_zlabel("")

# Title
ax.set_title("Energy Flow", pad=20)

plt.show()
```

## ðŸŒ The Greenhouse Effect (Natural vs. Anthropogenic)

The **greenhouse effect** is a natural physical process that keeps Earth warm enough to support life.  
However, human activities have **intensified** this effect, leading to global warming.

---

### ðŸ”¬ Mechanism

1. The Sun emits **short-wave radiation** that passes through the atmosphere.  
2. Earthâ€™s surface **absorbs** this energy and warms up.  
3. The planet then **reemits energy as long-wave infrared radiation**.  
4. **Greenhouse gases (GHGs)** trap part of this infrared radiation, preventing heat from escaping into space.

ðŸ‘‰ This natural heat retention makes Earth habitable.  
ðŸ‘‰ Extra greenhouse gases from human activity **enhance** the effect and increase global temperature.

---

### ðŸŒ± Natural vs. Anthropogenic Greenhouse Effect

**Natural greenhouse effect**
- Maintains Earthâ€™s average temperature around **15 Â°C**.  
- Essential for life on the planet.

**Anthropogenic (human-caused) greenhouse effect**
- Produced by burning fossil fuels, deforestation, and industry.  
- Increases the concentration of greenhouse gases.  
- Leads to **global warming and climate change**.

---

### ðŸ§ª Key Greenhouse Gases

- **Carbon dioxide ($CO_2$)** â†’ Most important by **total volume emitted**.  
- **Methane ($CH_4$)** â†’ Much **higher warming potential** than $CO_2$, but lower concentration.  
- **Nitrous oxide ($N_2O$)** â†’ Strong and long-lasting warming effect.  
- **Fluorinated gases** â†’ Synthetic gases with **very high warming potential**.

---

## ðŸ§­ Simple Idea

**Sunlight enters â†’ Earth absorbs â†’ Heat tries to escape â†’ Gases trap part of it â†’ Planet warms**

Human activity makes this **natural blanket thicker**.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# --- Figure setup ---
fig = plt.figure(figsize=(9, 7))
ax = fig.add_subplot(111, projection="3d")

# --- Create Earth sphere ---
u = np.linspace(0, 2 * np.pi, 200)
v = np.linspace(0, np.pi, 200)
r_earth = 1.0

x = r_earth * np.outer(np.cos(u), np.sin(v))
y = r_earth * np.outer(np.sin(u), np.sin(v))
z = r_earth * np.outer(np.ones_like(u), np.cos(v))

# Earth surface
ax.plot_surface(x, y, z, linewidth=0, antialiased=True, shade=True)

# --- Atmosphere shell (greenhouse gases layer) ---
r_atm = 1.15
x2 = r_atm * np.outer(np.cos(u), np.sin(v))
y2 = r_atm * np.outer(np.sin(u), np.sin(v))
z2 = r_atm * np.outer(np.ones_like(u), np.cos(v))

ax.plot_surface(x2, y2, z2, linewidth=0, alpha=0.15)

# --- Incoming solar radiation (parallel rays) ---
sun_rays_start = [-3, -3, -3]
for z0 in [-0.8, 0, 0.8]:
    ax.quiver(-3, 0, z0, 2.2, 0, -0.2*z0, arrow_length_ratio=0.08)

ax.text(-3.2, 0, 1.4, "Incoming solar radiation\n(short-wave)", ha="center")

# --- Outgoing infrared radiation ---
for angle in np.linspace(-40, 40, 5):
    dx = 1.4 * np.cos(np.radians(angle))
    dz = 1.4 * np.sin(np.radians(angle))
    ax.quiver(0, 0, 1, dx, 0, dz, arrow_length_ratio=0.08)

ax.text(1.8, 0, 1.6, "Outgoing infrared\n(long-wave)", ha="center")

# --- Trapped heat arrows (back toward Earth) ---
for angle in [-30, 0, 30]:
    dx = -0.6 * np.cos(np.radians(angle))
    dz = -0.6 * np.sin(np.radians(angle))
    ax.quiver(0.8*np.cos(np.radians(angle)), 0, 1.05 + 0.3*np.sin(np.radians(angle)),
              dx, 0, dz, arrow_length_ratio=0.1)

ax.text(0, 0, 1.9, "Heat partially trapped\nby greenhouse gases", ha="center")

# --- Formatting for clean scientific look ---
ax.set_box_aspect([1, 1, 1])
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])
ax.set_xlabel("")
ax.set_ylabel("")
ax.set_zlabel("")
ax.set_title("Greenhouse Effect Mechanism", pad=20)

plt.show()
```

## âš–ï¸ Energy Balance and Units

Understanding energy systems requires distinguishing between **power** and **energy**, as well as knowing how energy is **released or absorbed** in chemical processes.

---

### ðŸ”Œ Power

**Power** measures how fast energy is transferred or used.  
It represents an **instantaneous flow**.

**Common units:**
- Watt (**W**)
- Kilowatt (**kW**) = 1,000 W
- Megawatt (**MW**) = 1,000,000 W

ðŸ‘‰ Think of power as the **speed of energy use**.

---

### ðŸ”‹ Energy

**Energy** is the total amount accumulated over time.  
It represents a **stored quantity**.

**Common units:**
- Watt-hour (**Wh**)
- Kilowatt-hour (**kWh**)
- Joule (**J**) â†’ SI unit of energy

ðŸ‘‰ Think of energy as the **size of the tank**, while power is the **flow rate**.

---

### ðŸ§ª Energy in Chemical Bonds

When chemical reactions occur:

- **Breaking bonds** â†’ **absorbs energy**  
- **Forming bonds** â†’ **releases energy**

Example: Formation of water from hydrogen and oxygen:

$$
2H_2 + O_2 \rightarrow 2H_2O
$$

This reaction **releases energy** because the new **Oâ€“H bonds** are more stable than the original bonds.

This released energy is related to the **bond enthalpy** and is the basis of many **energy technologies**.

---

## ðŸ§­ Key Idea

**Power = rate of energy flow**  
**Energy = accumulated amount**

Chemical reactions transform **stored chemical energy** into **usable energy**.


```python
import numpy as np
import matplotlib.pyplot as plt

# Reaction coordinate
x = np.linspace(0, 10, 400)

# Smooth exothermic energy curve
y = 4 * np.exp(-0.5 * (x - 4)**2) + 2 - 0.15 * x

reactants_energy = y[0]
products_energy = y[-1]
peak_energy = max(y)

# Create figure
plt.figure(figsize=(8, 5))

# Plot curve
plt.plot(x, y, linewidth=2)

# Reference dashed lines
plt.hlines(reactants_energy, 0, 2, linestyles="dashed")
plt.hlines(products_energy, 8, 10, linestyles="dashed")

# Reactants & products labels (moved for clarity)
plt.text(0.2, reactants_energy + 0.25, "Reactants", fontsize=11)
plt.text(8.2, products_energy - 0.45, "Products", fontsize=11)

# Î”H vertical arrow
plt.annotate(
    "",
    xy=(9.3, products_energy),
    xytext=(9.3, reactants_energy),
    arrowprops=dict(arrowstyle="<->", linewidth=1.5)
)

plt.text(
    9.4,
    (reactants_energy + products_energy) / 2,
    "Î”H",
    va="center",
    fontsize=12
)

# Bond breaking / forming annotations (repositioned higher & lower)
plt.text(3.8, peak_energy + 0.6, "Bond breaking\n(energy absorbed)", ha="center")
plt.text(6.8, products_energy + 1.0, "Bond forming\n(energy released)", ha="center")

# Axis labels
plt.xlabel("Reaction Progress")
plt.ylabel("Energy")

# Add vertical padding so nothing overlaps the title
plt.ylim(products_energy - 1.2, peak_energy + 1.2)

# Clean academic style
plt.grid(True)
for spine in ["top", "right"]:
    plt.gca().spines[spine].set_visible(False)

# Title with extra padding
plt.title("Energy Profile of an Exothermic Reaction", pad=18)

plt.tight_layout()
plt.show()
```

## ðŸŒ Goals of the Paris Agreement

The **Paris Agreement** is a global climate treaty aimed at reducing the risks of **climate change** by limiting the rise in Earth's average temperature.

---

### ðŸŒ¡ï¸ Temperature Limits

Countries agreed to:

- Keep the global temperature increase **well below 2 Â°C** above **pre-industrial levels**.  
- **Preferably limit warming to 1.5 Â°C**, since this significantly reduces:
  - Extreme weather events  
  - Sea-level rise  
  - Ecosystem damage  

These limits define the **political and scientific targets** for climate action.

---

### ðŸ§® Carbon Budget

The **carbon budget** represents:

ðŸ‘‰ The **maximum amount of COâ‚‚** humanity can still emit  
before exceeding the **1.5 Â°C or 2 Â°C warming limits**.

Key ideas:

- More emissions today â†’ **less budget remaining**.  
- Reaching **net-zero emissions** stops the budget from shrinking.  
- The carbon budget connects:
  - **Science** (climate physics)  
  - **Policy** (international agreements)  
  - **Energy transition** (renewables, efficiency, electrification)

---

## ðŸ§­ Core Concept

**Temperature limit â†’ defines total allowable COâ‚‚ â†’ determines global climate action.**


```python
import numpy as np
import matplotlib.pyplot as plt

# Cumulative CO2 emissions (arbitrary scale for conceptual diagram)
x = np.linspace(0, 100, 200)

# Temperature response curve (increasing with emissions)
temp = 0.02 * x

plt.figure(figsize=(8, 5))

# Temperature curve
plt.plot(x, temp, linewidth=2)

# Paris Agreement limits
plt.axhline(1.5, linestyle="dashed")
plt.axhline(2.0, linestyle="dashed")

# Carbon budget intersection points
x_15 = 1.5 / 0.02
x_20 = 2.0 / 0.02

plt.axvline(x_15, linestyle="dotted")
plt.axvline(x_20, linestyle="dotted")

# Labels (carefully positioned to avoid overlap)
plt.text(2, 2.15, "Temperature response to cumulative COâ‚‚", fontsize=11)

plt.text(102, 1.5, "1.5 Â°C limit", va="center")
plt.text(102, 2.0, "2 Â°C limit", va="center")

plt.text(x_15, -0.15, "Remaining carbon budget\nfor 1.5 Â°C", ha="center")
plt.text(x_20, -0.35, "Remaining carbon budget\nfor 2 Â°C", ha="center")

# Axes labels
plt.xlabel("Cumulative COâ‚‚ Emissions")
plt.ylabel("Global Temperature Increase (Â°C)")

# Limits and spacing to prevent overlaps
plt.xlim(0, 110)
plt.ylim(-0.5, 2.4)

# Clean academic style
plt.grid(True)
for spine in ["top", "right"]:
    plt.gca().spines[spine].set_visible(False)

plt.title("Paris Agreement: Temperature Limits and Carbon Budget", pad=18)

plt.tight_layout()
plt.show()
```
